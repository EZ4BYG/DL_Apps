```python
img_path = glob.glob( r'D:\SGDownload\611_早上\未命名(2)\未命名(2)_大图\待处理png图\4图合并测试\*' )
img_path = sorted(img_path, key = lambda x:int(x.split('\\')[-1].split('.')[0].split('_')[-1]) ) 
```

```python
def img_recover(img_path):
    total = len(img_path)
    img = [] 
    for x in range(total):
        tmp = tf.io.read_file( img_path[x] )
        tmp = tf.image.decode_png( tmp, channels = 3 )
        tmp = tmp.numpy().reshape(256,256,3)
        img.append( tmp )
        
    c1 = img[0:2]
    c2 = img[2:4]
    
    c11 = np.concatenate( c1, axis = 1 )
    c22 = np.concatenate( c2, axis = 1 )
    c = np.concatenate( [c11,c22], axis = 0 )
    
    return c
```

```python
def label_recover(label_path):
    total = len(label_path)
    img = [] 
    for x in range(total):
        tmp = tf.io.read_file( label_path[x] )
        tmp = tf.image.decode_png( tmp, channels = 1 )
        tmp = tmp.numpy().reshape(256,256)
        img.append( tmp )
        
    c1 = img[0:2]
    c2 = img[2:4]
    
    c11 = np.concatenate( c1, axis = 1 )
    c22 = np.concatenate( c2, axis = 1 )
    c = np.concatenate( [c11,c22], axis = 0 )
    
    return c
```

```python
img_path =  glob.glob( r'D:\SGDownload\611_早上\未命名(2)\未命名(2)_大图\待处理png图\4图合并测试\图像\*' )
label_path =  glob.glob( r'D:\SGDownload\611_早上\未命名(2)\未命名(2)_大图\待处理png图\4图合并测试\标签\*' )

img = img_recover(img_path)
label = label_recover(label_path)
```

```python
plt.figure( )
plt.subplot(1,2,1)
plt.imshow(img)
plt.subplot(1,2,2)
plt.imshow(label)
```

```python
def img_label_save(img, label):
    # 保存地址：
    path_img = 'D:/SGDownload/611_早上/未命名(2)/未命名(2)_大图/待处理png图/4图合并测试/图像/'
    path_label = 'D:/SGDownload/611_早上/未命名(2)/未命名(2)_大图/待处理png图/4图合并测试/标签/'
    # 保存：
    scipy.misc.imsave( path_img + 'total_img.png', img )
    scipy.misc.imsave( path_label + 'total_label.png', label )
```
