```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
import matplotlib.image as mpimg # mpimg 用于读取图片
```

```python
# 读取所有图片：
# 左边3个：
left1 = mpimg.imread('376.png')
left2 = mpimg.imread('382.png')
left3 = mpimg.imread('383.png')
# 中间3个：
center1 = mpimg.imread('386.png')
center2 = mpimg.imread('402.png')
center3 = mpimg.imread('582.png')
# 右边3个：
right1 = mpimg.imread('536.png')
right2 = mpimg.imread('543.png')
right3 = mpimg.imread('551.png')

all_img = [left1, center1, right1, left2, center2, right2, left3, center3, right3]
```

```python
title = ['子图1', '子图2','子图3']
```

```python
plt.figure( figsize = (16,9) )
for i in range(1, 10):
    ax = plt.subplot(3, 3, i)
    plt.imshow( all_img[i-1] )
    
    # 去除坐标轴
    plt.xticks([])
    plt.yticks([])
    # 去除黑框
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.spines['bottom'].set_visible(False)
    ax.spines['left'].set_visible(False)    


plt.subplots_adjust(hspace = 0.1, wspace = 0.2)
# plt.subplots_adjust(left = 0.01, top = 0.95, right = 0.95, bottom = 0.01)
# plt.subplots_adjust( wspace = 0.05, hspace = 0.005 )
# plt.subplots_adjust( hspace = 0.01, wspace = 0.01 )
plt.tight_layout()
plt.savefig('all.jpg')
plt.show()
```

```python
z1 = [left1, left2, left3]
z2 = [center1, center2, center3]
z3 = [right1, right2, right3]

for i in range(1, 4):
    ax = plt.subplot(3, 1, i)
    plt.imshow( z1[i-1] )
    
    # 去除坐标轴
    plt.xticks([])
    plt.yticks([])
    # 去除黑框
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.spines['bottom'].set_visible(False)
    ax.spines['left'].set_visible(False)
    
plt.subplots_adjust(left = 0, right = 0, top = 0, bottom = 0, wspace=0.02, hspace=0.02)
plt.savefig('z1.png')
```
