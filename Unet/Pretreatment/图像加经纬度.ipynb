```python
img_path = glob.glob( r'D:\SGDownload\611_早上\未命名(2)\未命名(2)_大图\待处理png图\加经纬度子图\图像\*')
label_path = glob.glob( r'D:\SGDownload\611_早上\未命名(2)\未命名(2)_大图\待处理png图\加经纬度子图\标签\*' )
```

```python
# 读取：
img = []
for x in range( len(img_path) ):
    tmp = tf.io.read_file( img_path[x] )
    tmp = tf.image.decode_png(tmp, channels = 3)
    tmp = tmp.numpy()
    img.append( tmp )
    
label = [] 
for x in range( len(label_path) ):
    tmp = tf.io.read_file( label_path[x] )
    tmp = tf.image.decode_png(tmp, channels = 3)
    tmp = tmp.numpy()
    label.append( tmp )
```

```python
path = 'D:/SGDownload/611_早上/未命名(2)/未命名(2)_大图/待处理png图/加经纬度子图/合并/'
con = []
for x in range( len(img) ):
    tmp = np.concatenate( (img[x],label[x]), axis = 1 )
    con.append(tmp)
    plt.imsave( path + '{}.png'.format(x), tmp)
```
