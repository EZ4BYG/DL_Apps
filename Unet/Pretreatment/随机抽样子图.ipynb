```python
import tensorflow as tf
import numpy as np
import glob
import matplotlib.pyplot as plt
import matplotlib                 # 文件保存为png + 文件读取为png
from collections import Counter  # 数组独特元素统计
import scipy
import random
```

```python
all_image_path = glob.glob( r'D:\SGDownload\611_早上\未命名(2)\未命名(2)_大图\待处理png图\大图分割后的png\*' )
all_image_path = sorted(all_image_path, key = lambda x:int(x.split('\\')[-1].split('.')[0].split('_')[-1]) )[100:]
total = len(all_image_path)
print( total )
all_image_path[0:5]
```

```python

# 在给定范围内(索引值)，随机抽10个数：
select_total = 500
index = random.sample( range(0,total),select_total )
list.sort( index )  # 排个升序
index
```

```python
image_select_path = []  # 保存按索引值提取的文件的“路径”
for x in index:
    image_select_path.append( all_image_path[x] )
```

```python
len(image_select_path)
```

```python
# 读取文件 + 转存到新地址：
save_path = 'D:/SGDownload/611_早上/未命名(2)/未命名(2)_大图/待处理png图/测试数据集/图像/'

def read_save(image_select_path, index):
    for x in range( len(image_select_path) ):
        # 读取：
        img = tf.io.read_file( image_select_path[x] )  # 记得是列表，必须加[0]
        img = tf.image.decode_png(img, channels = 3)
        img = img.numpy()
        # 转存：
        matplotlib.image.imsave( save_path + 'subimg_{}.png'.format(index[x]+100), img )
```

```python
read_save(image_select_path, index)
```
